```python
import pandas as pd
import numpy as np
import tensorflow as tf
import plotly
from plotly.express import scatter
```

```python
data = pd.read_csv('./datasets/diabetes.csv')

data.head()
```

```python
X = data.iloc[:, 1:-1]
y = data.iloc[:, -1]
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
y = pd.get_dummies(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)
```

```python
def NurNet(norons_in_layer, epochs, batch_size):
    Layers_num = len(norons_in_layer)
    # str = 'tf.keras.models.Sequential([tf.keras.layers.Dense(norons_in_layers[0], activation="relu", input_shape=(7, ))])'
    str="tf.keras.models.Sequential([tf.keras.layers.Dense(norons_in_layer[0], activation='relu', input_shape=(7,)),"

    for i in range(1, Layers_num):
        # str = str+f"tf.keras.layers.Dense({norons_in_layer[i]}, activation='relu')"
        # str = str+f"tf.keras.layers.Dense(2, actovation='softmax')"
        str=str+f"tf.keras.layers.Dense({norons_in_layer[i]}, activation='relu'),"
        str=str+f"tf.keras.layers.Dense(2, activation='softmax')])"

        model = eval(str)
        # model.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss=tf.keras.losses.CategorycalCrossentropy(), metrics=[tf.keras.metrics.CategoricalCrossentropy(), tf.keras.metrics.Recall(class_id=1)])
        model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
                  loss=tf.keras.losses.CategoricalCrossentropy(),
                  metrics=[tf.keras.metrics.CategoricalAccuracy(),
                           tf.keras.metrics.Recall(class_id=1)]
                 )
        history = model.fit(X_train, y_train, epochs=epochs, validation_data=(X_test, y_test), batch_size=batch_size)
        keys = list(history.history.keys())

        accuracy_train = np.max(history.history['categorical_accuracy'])
        accuracy_test = np.max(history.history[keys[2]])
        recall_train = np.max(history.history[keys[4]])
        recall_test = np.max(history.history[keys[5]])

        return (model, accuracy_train, accuracy_test, recall_train, recall_test)
    

NurNet([25, 34, 45, 36, 18, 12, 7], 50, 50)
```

```python
results = pd.DataFrame(columns=['Layers_num', 'norons_in_layer', 'model', 'accuracy_train', 'accuracy_test', 'recall_train', 'recall_test'])
architecture_num = 0
```

```python
batch_size = 50
epochs = 50

for layers_num in range(3, 10):
    for architecture in range(5):
        norons_in_layers = np.random.randint(5, 50, size=layers_num)
        model, bst_acc_train, bst_acc_test, bst_rec_train, bst_rec_test = NurNet(norons_in_layers, epochs, batch_size)
        results.loc[architecture_num]= [layers_num, str(norons_in_layers), model, bst_rec_train, bst_rec_test, bst_rec_train, bst_rec_test]
        architecture_num+=1
```
